```python
import numpy as np
import random
import scipy
import pickle
import os
import matplotlib.pyplot as plt
from sklearn import manifold, datasets, decomposition, ensemble, random_projection
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

load the data

```python
if os.path.isfile('mnist_X.pickle') and os.path.isfile('mnist_Y.pickle'):
    mnist_X = pickle.load(open('mnist_X.pickle', 'rb'))
    mnist_Y = pickle.load(open('mnist_Y.pickle', 'rb'))
    mnist_X_mini = pickle.load(open('mnist_X_mini.pickle', 'rb'))
    mnist_Y_mini = pickle.load(open('mnist_Y_mini.pickle', 'rb'))
    mnist_X_tsne = pickle.load(open('mnist_X_tsne.pickle', 'rb'))
else:
    mnist = input_data.read_data_sets("MNIST_data/", reshape=True)
    # notice that here we don't shuffle the data so the samples with same label are together
    sample_per_label = 1000
    random.seed(291)
    index_sampling = []
    for target_label in range(10):
        target_label_position = np.where(mnist.train.labels==target_label)[0]
        target_label_sampling = random.sample(list(target_label_position), sample_per_label)
        index_sampling.extend(target_label_sampling)
    mnist_X = mnist.train.images[index_sampling,:]
    mnist_Y = mnist.train.labels[index_sampling]
```

```python
plt.imshow(np.reshape(mnist_X[5000,:],[28,28]))
plt.show()
```

calculate the distance matrix

```python
if os.path.isfile('mnist_Dist.pickle'):
    mnist_Dist = pickle.load(open('mnist_Dist.pickle', 'rb'))
else:
    mnist_Dist = scipy.spatial.distance.cdist(mnist_X,mnist_X)
```

```python
def plot_embedding_2d(X, title=None):
    #坐标缩放到[0,1]区间
    x_min, x_max = np.min(X,axis=0), np.max(X,axis=0)
    X = (X - x_min) / (x_max - x_min)

    #降维后的坐标为（X[i, 0], X[i, 1]），在该位置画出对应的digits
    fig = plt.figure(figsize=[10,10])
    ax = fig.add_subplot(1, 1, 1)
    for i in range(X.shape[0]):
        ax.text(X[i, 0], X[i, 1],str(mnist_Y[i]),
                 color=plt.cm.Set1(mnist_Y[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    if title is not None:
        plt.title(title)
```

```python
X_pca = decomposition.TruncatedSVD(n_components=2).fit_transform(mnist_X)
plot_embedding_2d(X_pca[:,0:2],"PCA 2D")
```

```python
tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
X_tsne = tsne.fit_transform(mnist_X)
print(X_tsne.shape)
plot_embedding_2d(X_tsne[:,0:2],"t-SNE 2D")
```

```python
N = 10000

# X = tf.placeholder('float')
X = tf.placeholder(tf.float32,[None,784])
Xshape = tf.shape(X)
 
sqX = tf.reduce_sum(X*X, 1)
sqX = tf.reshape(sqX, [-1, 1])
sqDX = sqX - 2*tf.matmul(X, tf.transpose(X)) + tf.transpose(sqX)
# sqDXarray = tf.stack([sqDX[i, j] for i in range(N) for j in range(i+1, N)])
sqDXarray = tf.add(sqDX,tf.eye(N))
DXarray = tf.sqrt(sqDXarray)
DXarray = tf.reshape(DXarray,[-1,1]) 

# Y = tf.Variable(init_points, dtype='float')
# Y = tf.Variable(mnist_X_tsne[:N,:2], dtype='float')
# Y = tf.random_normal_initializer([N,2],stddev = 1.0)
#Y = tf.layers.dense(Y, 64, use_bias=False, activation=None)
#Y = tf.nn.relu(Y)
#Y = tf.layers.dense(Y, 2, use_bias=False, activation=None)
Y = tf.get_variable("Y", [N, 2], dtype=tf.float32, initializer=tf.random_normal_initializer(stddev = 5))
sqY = tf.reduce_sum(Y*Y, 1)
sqY = tf.reshape(sqY, [-1, 1])
sqDY = sqY - 2*tf.matmul(Y, tf.transpose(Y)) + tf.transpose(sqY)
# sqDYarray = tf.stack([sqDY[i, j] for i in range(N) for j in range(i+1, N)])
sqDYarray = tf.add(sqDY,tf.eye(N))
DYarray = tf.sqrt(sqDYarray)
DYarray = tf.reshape(DYarray,[-1,1]) 
```

```python
Z = (tf.reduce_sum(DXarray)-N)*0.5
avg_dist = tf.divide(tf.reduce_sum(tf.abs(DXarray-DYarray)),(N**2-N)/2)
diffXY = DXarray-DYarray
divider = tf.reduce_min(DXarray)
numerator = tf.reduce_sum(tf.divide(tf.square(DXarray-DYarray), DXarray))
cost = tf.reduce_mean(tf.square(DXarray-DYarray))
# cost = tf.divide(numerator, Z)
```

```python
# update_rule = tf.assign(Y, Y-0.01*grad_cost/lapl_cost)
train = tf.train.GradientDescentOptimizer(0.0000001).minimize(cost)
# train = tf.train.AdamOptimizer(0.000001).minimize(cost)
# train = tf.train.RMSPropOptimizer(0.000001).minimize(cost)
init = tf.global_variables_initializer()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
nbsteps = 100

for i in range(nbsteps):
    #sess.run(train, feed_dict={X: mnist_X_mini})
    #c,a_dist,nume,divid = sess.run([cost,avg_dist,numerator,divider], feed_dict={X: mnist_X_mini})    
       
    sess.run(train, feed_dict={X: mnist_X[:N,:]})
    c,a_dist,DY,differ = sess.run([cost,avg_dist,DYarray,diffXY], feed_dict={X: mnist_X[:N,:]})    
    
    #print(sess.run(DYarray, feed_dict={X: mnist_X[:N,:]}))
    if i%1 == 0:
        print("epoch: ", i, " cost = ", c, " avg_dist: ", a_dist, DY,differ)
embedding = sess.run(Y, feed_dict={X: mnist_X[:N,:]})
```
